```python
#import pandas library
import pandas as pd
```

```python
# read dataset
df = pd.read_csv('/home/ls283h/Work/EcoQuartier/iris_in_qpv.csv')
df = pd.concat([df.loc[:,['area_qpv','fid_qpv']],df.loc[:,'NUMPOINTS_ETI':]], axis=1)
df
```

```python
# multiply proportion of Iris area in EQ area by variables of interest 
df1 = df[df.columns[2:-1]].multiply(df.area_iris_in_qpv/df.area_qpv, axis=0)
```

```python
# concatenate iris id again
df2 = pd.concat([df['fid_qpv'], df1], axis=1)
df2
```

confirmed with Hugo:NA values are replace by zero before the sum (min_count can be set to 1) but if there is only a missing value in the group then the result will be NaN!

```python
# final weighted mean by summing up weighted values
df_weighted = df2.groupby('fid_eq').sum(min_count=1)
df_weighted
```

```python
# identify datazones with missing values
df_weighted[df_weighted.isna().any(axis=1)]
```

```python
# saving dataframe to csv
df_weighted.to_csv('/home/ls283h/Work/EcoQuartier/eq_with_weighted_iris.csv', index=True)
```
