```python
# open a file for writing and create a csv writer object
with os.scandir('/home/ls283h/Work/CCTV/xml/') as entries:
    for entry in entries:
        if entry.is_file() and entry.name.endswith('xml'):
            tree = ET.parse(entry)
            root = tree.getroot()

            if os.path.isfile('/home/ls283h/Work/CCTV/cctv_manual_counts.csv'):
                with open('cctv_manual_counts.csv', 'a') as file:
                    writer = csv.writer(file, delimiter=",", lineterminator="\n")

                    image_name = root.find('filename').text
                    image_id = image_name[:17]
                    timestamp_utc = image_name[0:13]
                    timestamp_utc = pd.to_datetime(timestamp_utc, unit='ms')
                    timestamp_utc = timestamp_utc.floor('min')
                    date = timestamp_utc.strftime('%Y-%m-%d')
                    time = timestamp_utc.strftime('%H:%M')
                    camera = image_name[14:17]

                    for member in root.findall('object'):
                        row = [image_id, date, time, camera, member.find('./name').text, 
                               member.find('./bndbox/xmin').text, member.find('./bndbox/ymin').text]
                        writer.writerow(row)
            else:
                with open('/home/ls283h/Work/CCTV/cctv_manual_counts.csv', 'w') as file:
                    file.write('Image_id,Date,Time,Camera,Object,Pos_x,Pos_y\n')
                    writer = csv.writer(file, delimiter=",", lineterminator="\n")

                    image_name = root.find('filename').text
                    image_id = image_name[:17]
                    timestamp_utc = image_name[0:13]
                    timestamp_utc = pd.to_datetime(timestamp_utc, unit='ms')
                    timestamp_utc = timestamp_utc.floor('min')
                    date = timestamp_utc.strftime('%Y-%m-%d')
                    time = timestamp_utc.strftime('%H:%M')
                    camera = image_name[14:17]

                    for member in root.findall('object'):
                        row = [image_id, date, time, camera, member.find('./name').text, 
                               member.find('./bndbox/xmin').text, member.find('./bndbox/ymin').text]
                        writer.writerow(row)
```
